```python
class seÃ±oron:
  def __init__(self,numero):
    self.num=numero

  def primario(self,numero):#numero=7
    vrai=False
    par=int(numero/2)
    resta=numero-(par*2)
    if resta==0:
      vrai=True
    return vrai

  def decisiones(self):
    gato=(self.num*3)+1
    while gato !=1:
      par=self.primario(gato)
      if  par==True:
        print("par")
        gato=gato/2
        print(gato,end="\n \n")
      else:
        print("impar")
        gato=(gato*3)+1
        print(gato,end="\n \n")
    return gato

master=seÃ±oron(0.5)
master.decisiones()
```
